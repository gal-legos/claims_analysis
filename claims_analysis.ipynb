### Q1: Provider Analysis
Used the HEADER file to find the top 5 billing providers by number of claims. To do this, we first had to merge the Billing Provider names into one column and then utilize the sort_values() function and groupby function to get the top 5 billing providers. We see that SB Internists has the highest number of claims, with 152 claims, and SB Psychiatric Associates comes in at number 5 with 36 claims.

```python
df_header["BillingProviderName"] = (
    df_header["BillingProvFirstName"].astype(str) + " " +
    df_header["BillingProvLastName"].astype(str)
)


provider_counts = df_header.groupby(["BillingProviderName", "BillingProviderNPI"]).size().reset_index(name="ClaimCount")
top5_providers = provider_counts.sort_values("ClaimCount", ascending=False).head(5)
display(top5_providers)

# Bar chart
top5_providers.plot(x="BillingProviderName", y="ClaimCount", kind="bar")
plt.title("Top 5 Providers by Claim Volume")
plt.show()
```

### Q4: Common Procedures
For this question, the LINE file was used to find the top 10 most common HCPCS codes billed within a claim period. Displayed below, we can see that the HCPCS code 99291 was billed the most, with 68 times. Within the code script, the ClientProcedureName was included to give us the description of the HCPCS codes. As we can see from the description, this is a very important code, and that makes sense as to why it was billed the most.

```python
# Count how often each HCPCS appears
hcpcs_counts = df_line["HCPCS"].value_counts().reset_index()
hcpcs_counts.columns = ["HCPCS", "Count"]

# Get a unique mapping of HCPCS  ClientProcedureName
hcpcs_desc = df_line[["HCPCS", "ClientProcedureName"]].drop_duplicates()

# Merge counts with descriptions
hcpcs_with_desc = hcpcs_counts.merge(hcpcs_desc, on="HCPCS", how="left")

# Take the top 10
top10_procedures = hcpcs_with_desc.head(10)

display(top10_procedures)

# Bar chart
top10_procedures.plot(kind="bar")
plt.title("Top 10 Billed Procedures")
plt.show()
```

### Q5: Service Location Analysis
For this question, the HEADER file was utilized once again, but this time to find the number of claims submitted for each Place of Service. Once that number was found, we could then find the percentages of claims for each Place of Service and how the Doctor's Office and Inpatient claim percentages compare. From the table displayed, we can see that Inpatient claims reach about 60% while only 34% of the claims submitted were for Doctors' Offices. This seems logical considering that doctors' office visits are typically quick visits, and claims can range depending on the season, whereas inpatient claims will be more consistent due to round-the-clock care. 

```python
pos_facility_counts = (
    df_header.groupby(["PlaceOfService", "Facility_Location"])
    .size()
    .reset_index(name="Count")
)

display(pos_facility_counts)

total_claims = pos_facility_counts["Count"].sum()

pos_facility_counts["Percentage"] = (
    pos_facility_counts["Count"] / total_claims * 100
)

display(pos_facility_counts)
```

### Q6: Claims with High Service Line Counts
For this section of the project, two or more files were merged for analysis. Specifically, the HEADER and LINE files were combined to identify claims containing five or more service lines. Based on the final table shown below, only a small number of claims met this threshold; however, among those that did, Obstetrics and Child Services appeared most frequently as the departments billing in this manner.  

```python
# Look at the provider name columns first
print("Let's see what the provider columns look like:")
print(df_header[["BillingProvFirstName", "BillingProvLastName"]].head())

# Combine first and last name into one column
# Fill empty values with blank text, then combine them
df_header["BillingProviderName"] = df_header["BillingProvFirstName"].fillna("") + " " + df_header["BillingProvLastName"].fillna("")

# Remove extra spaces
df_header["BillingProviderName"] = df_header["BillingProviderName"].str.strip()


# Merge the two files together

merged_hl = df_header.merge(df_line, on="ProspectiveClaimId", how="left")

line_counts = merged_hl.groupby("ProspectiveClaimId").size().reset_index(name="NumLines")

high_line_claims = line_counts[line_counts["NumLines"] >= 5]
display(high_line_claims)

# Add provider + total charges
charges_per_claim = merged_hl.groupby("ProspectiveClaimId")["LinePos"].sum().reset_index(name="TotalCharges")

result = high_line_claims.merge(df_header[["ProspectiveClaimId", "BillingProviderName"]], on="ProspectiveClaimId")
result = result.merge(charges_per_claim, on="ProspectiveClaimId")

display(result.head())

```

### Q8: Charges by Payer
For question 8, the HEADER and LINE files were merged to calculate the total charges submitted per claim and then grouped by payer. The top 10 payers are shown in a table that includes their total charges, average charges per claim, and claim counts. From this summary, Medicare emerges as the payer with the highest total charges at 131,008. This result is expected, as Medicare typically accounts for a large share of healthcare utilization due to the size and continued growth of the elderly population in the United States

```python
claim_charges = df_line.groupby('ProspectiveClaimId')['Charges'].sum().reset_index()
claim_charges.columns = ['ProspectiveClaimId', 'TotalCharges']


print(claim_charges.head())
```

```python
# Select only the columns we need from HEADER
header_subset = df_header[['ProspectiveClaimId', 'PrimaryPayerName']]

# Combine the two datasets
merged_df = claim_charges.merge(header_subset, on='ProspectiveClaimId', how='left')


print(merged_df.head())
```

```python
payer_stats = merged_df.groupby('PrimaryPayerName').agg(
    TotalCharges=('TotalCharges', 'sum'),           # Sum of all charges
    AvgChargesPerClaim=('TotalCharges', 'mean'),    # Average charge per claim
    NumberOfClaims=('TotalCharges', 'count')        # Count of claims
).reset_index()


top_10_payers = payer_stats.sort_values('TotalCharges', ascending=False).head(10)


display(top_10_payers)
```

### Q9: Creative Analysis - Relationship Between Place of Service and Avg Charges

For Question 9, the HEADER and LINE files were used to visualize the relationship between Place of Service and Average Charges. Using the Python script, the files were merged, and the "Place of service", "Facility_Location", and "Charges" columns were selected to analyze the average charges per Place of service. The facility_location column was used as a description since the Place of Service only gives the code. From the Bar graph, we can see that ambulatory services have the highest average charges out of all the locations. This makes sense since a lot of patients are choosing ambulatory surgery recently, and doctors are also recommending ACS over inpatient, if possible. This also means that insurance companies are being charged more at Ambulatory service locations. 

```python
# Merge HEADER and LINE datasets on ProspectiveClaimId
merged_for_analysis = pd.merge(
    df_header,
    df_line,
    on="ProspectiveClaimId",
    how="left"
)

# Rename PlaceOfService_x to PlaceOfService because of the merge
if "PlaceOfService_x" in merged_for_analysis.columns:
    merged_for_analysis.rename(columns={"PlaceOfService_x": "PlaceOfService"}, inplace=True)

# Calculate average charges grouped by PlaceOfService and Facility_Location
avg_charges = (
    merged_for_analysis
    .groupby(["PlaceOfService", "Facility_Location"])["Charges"]
    .mean()
    .sort_values(ascending=False)
)

# Display results
display(avg_charges)

# Visualization
avg_charges.plot(kind="bar", figsize=(12,6))
plt.title("Average Charges by Place of Service and Facility Location")
plt.ylabel("Average Charges")
plt.xlabel("Place of Service & Facility Location")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

```
